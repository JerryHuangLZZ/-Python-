0D--标量  1D--向量  2D--矩阵  nD--多维空间

标准python不支持多维数组，我们需要引入numpy库：import numpy as np  (重命名为np简化拼写且避免了与其他库函数的冲突)

调用array()函数可创建多维数组ndarray对象

本章节大部分内容将用案例来引入相关操作

# 2.1 一维数组对象

```python
#例 2.1.1 创建两个一维数组，分别保存学生姓名与科目

import numpy as np
names=np.array(['Jerry','Zhou','Tom','Gangb'])
subjects=np.array(['数学','英语','Python','物理'])
print(names,subjects)
```

```python
#例 2.1.2 访问一维数组

print(names[2],'\n',names[-3])
```

```python
#例 2.1.3 一维数组切片，生成新数组（子集）

sub_subjects=subjects[0:4:2]
print(sub_subjects)
```

```python
#方法2 外层一对[]表示对数组元素的访问，内层一对[]是多个索引组成的列表

sub_subjects=subjects[[0,2,3]]
print(sub_subjects)
```

```python
#例 2.1.4 条件筛选出数组元素

find_subjects=subjects[(subjects=='数学')|(subjects=='Python')]
print(find_subjects)
```

```python
#实质：先用条件表达式创造出布尔型数组

bool_array=(subjects=='数学')|(subjects=='Python')
print(bool_array)
```

```python
#随后以此bool数组筛选元素，值为True选中，False不选

print(subjects[bool_array])
```

# 2.2 二维数组对象

```python
#例 2.2.1 创建二维数组scores，记录例2.1.1中各学生对应的各科成绩

scores = np.array([[70,85,77,90],[75,80,92,90],[90,93,88,87],[80,82,91,88]])
print(scores)
```

# 2.3 二维数组的相关操作

二维数组元素访问 格式：a[行，列]

二维数组切片 格式： a[行,列]，其中行列切片格式与一维数组相同

二维数组条件筛选：可以单独作用于行/列，也可以同时对行列进行筛选（分两层）

```python
#例 2.3.1 二维数组元素访问 

print(scores[2,1])
```

```python
print(scores[[1,3],[0,1]])  #逗号左边为行，右边为列，表示抽取位置为（1，0）与（3，1）的两个元素，生成一维数组
```

```python
#例 2.3.2 二维数组切片出特定列

print(scores[:,[0]])
```

```python
#例 2.3.3 二维数组切片出特定行

print(scores[[0,1,3],:])
```

```python
#可以省去列索引的：号

print(scores[[0,1,3]])  
```

```python
#例 2.3.4 二维数组切片出特定行与特定列

#先切片出特定行
row_scores=scores[[1,3]]
#再切片出特定列
row_column_scores=row_scores[:,[0,1,3]]

print(row_column_scores)
```

```python
#等同于两层切片

print(scores[[1,3]][:,[0,1,3]])
```

```python
#当列为连续的列，无此烦恼

print(scores[[1,3],0:3])
```

```python
#例 2.3.5 条件筛选二维数组(注：scores中names为行，subjects为列)

#单独作用于行

find_scores1=scores[(names=='Jerry')|(names=='Zhou'),:]   
print(find_scores1)
```

```python
#单独作用于列

find_scores2=scores[:,(subjects=='Python')|(subjects=='数学')]
print(find_scores2)
```

```python
#同时筛选

find_scores3=scores[(names=='Jerry')|(names=='Zhou')][:,(subjects=='Python')|(subjects=='数学')]  
print(find_scores)
```

# 2.5  多维数组运算

整体运算：广播机制

选取特定元素运算：切片得到元素再运算

```python
#例 2.5.1 二维数组与标量运算

scores+=5
print(scores)
```

```python
#例 2.5.2 二维数组与一维数组运算

addition=np.array([1,4,6,8])
new_scores=scores+addition
print(new_scores)
```

```python
#例 2.5.3 二维数组特定元素运算

Jerry_Python_score=scores[names=='Jerry',subjects=='Python']
Jerry_Python_score+=5
print(Jerry_Python_score)
```

# 2.6  通用函数

分为一元函数（1输入1输出，图1）与二元函数（2输入，1输出，图2）

对各个元素作用

```python
#例 2.6.1 使用subtract函数给每个学生分数减去10分

scores = np.array([[70,85,77,90],[75,80,92,90],[90,93,88,87],[80,82,91,88]])

new_score=np.subtract(scores,10)
print(new_score)
```

# 2.7  聚合函数

对行或列或全体元素作用

参数axis=0: 列，对同一列不同行的元素
参数axis=1: 行，对同一行不同列的元素

```python
#例 2.7.1 统计Jerry所有课程成绩的总分与平均分

sum_scores=scores[names=='Jerry',:].sum()
average_scores=scores[names=='Jerry',:].mean()
print(sum_scores,average_scores)
```

```python
#例 2.7.2 统计所有科目的平均分

average_scores=scores.mean(axis=0)
print(average_scores)
```

random模块提供了随机数生成函数

```python
#例 2.8.1 生成3x4的二维数组，元素为0～5的随机整数

rand=np.random.randint(0,6,size=(3,4))
print(rand)
```
