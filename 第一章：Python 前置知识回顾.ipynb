# 1.1 元组（Tuple）

一维的，有序的，不可修改的，可存储任意数据类型（包括元组）的 数据序列

用一对（）将所有元素括起来，使用 " 变量名[索引] "的格式访问，正序索引范围[0,n-1],逆序[-n,-1]

```python
#例 1.1.1 

yuanzu=('Jerry',26,('Calculus',4.0))
print(yuanzu)
```

```python
print(yuanzu[2])
```

```python
print(yuanzu[-3])
```

```python
print(yuanzu[2][1])
```

# 1.2 列表（List）

一维的，有序的，可修改的，可存储任意数据类型（包括列表）的 数据序列

用一对[]将所有元素括起来，使用 " 变量名[索引] "的格式访问，正序索引范围[0,n-1],逆序[-n,-1]

```python
#例 1.2.1

liebiao=['Tom',14,['Physics',1.0]]
print(liebiao)
```

```python
print(liebiao[1])
```

```python
print(liebiao[-2])
```

```python
print(liebiao[2][0])
```

# 1.3 元组与列表的相关操作

切片：选取数据序列的子集

按步长选取子集，a[i] 或 a[i:j] 或 a[i:j:k]

```python
#例 1.3.1

list=(1,21,4,5,99,'hello',('Jerry','handsome'))
print(list[1])
```

```python
# [i：j] 从i到j-1

print(list[1:4])
```

```python
#[i:j:k] 从i到j-1,步长为2

print(list[0:6:2])
```

# 1.4 字典（Dictionary）

一维的，无序的，可修改的，由“键值对”元素组成的 数据序列

用一对{}将所有元素括起来，用:分割“键”和”值“，用,隔开不同的键值对

“键” 为不可变对象，如 数，字符串，元组 ，通过“键”可以找到相对应的“值”

```python
#例 1.4.1

dictionary={'电路分析':99,'数电':100,'模电':(100,4.0)}
print(dictionary)
```

```python
print(dictionary['电路分析'])
```

```python
print(dictionary['模电'][1])
```

# 1.5  字典的相关操作

可以通过“键”对字典进行添加，删除，修改，遍历等操作

```python
#例 1.5.1

dictionary={'电路分析':99,'数电':100,'模电':(100,4.0)}
print(dictionary)
```

```python
#修改
dictionary['数电']=0
print(dictionary)
```

```python
#添加
dictionary['电子电路创新实践']=100
print(dictionary)
```

```python
#删除
del dictionary['电路分析']
print(dictionary)
```

```python
#遍历
dictionary={'电路分析':99,'数电':100,'模电':(100,4.0)}
for i in dictionary:
    print(i,':',dictionary[i])
```
