类的方法定义格式：Class=类名:
        
        xxx_1=None/要记录的东西
        xxx_2=None/要记录的东西
        def 方法名（self,形式参数1，形式参数2，...）：

```python
#例1.3.1 定义类Student，较之前添加一个方法，实现对学号和年龄的输出

class Student:   
    name=None
    number=None
    age=None
    def method1(self):
        print(self.number,self.age)

jerry=Student()   
jerry.name="jerry"
jerry.number=666
jerry.age=18

jerry.method1()
```

```python
#例1.3.2 用到形式参数的情形：添加方法，实现对名字与传入的形式参数的输出

class Student:   
    name=None
    number=None
    age=None
    def method2(self,a):
        print(self.name,a)

jerry=Student()   
jerry.name="jerry"
jerry.number=666
jerry.age=18

jerry.method2("love Zhou")
```

# 1.4 创造对象的构建函数：Python 中，如果类的每一个对象都需要完成相同的初始化工作，则可以定义 __init__函数，使得创建类的对象时可以自动执行该函数

格式： def __init__(self,形式参数1,....):    (init两侧分别是两个下划线)

其中一个应用：可以一次性赋值多个数据对象


```python
#例1.4.1  实现例1.2.2的更高效方法（一次性填入数据） 

class Student:   
    name=None
    number=None
    age=None
    def __init__(self,a,b,c):
        self.name=a
        self.number=b
        self.age=c

jerry=Student("jerry",666,18)

print(jerry.name,jerry.number,jerry.age)
```

# 1.5 单继承：在原先类的基础上做修补形成新的类

格式：class 子类名（父类名）：
           
           增添内容

```python
#例1.5.1 给类Student增添家乡信息与表白功能

#父类
class Student:   
    name=None
    number=None
    age=None
    def __init__(self,a,b,c):
        self.name=a
        self.number=b
        self.age=c

#子类
class Student_2(Student):
    hometown=None
    def lovezz(self):
        return("jerrylovezhouz")
       
zhou=Student_2("zhou",999,18)
zhou.hometown='Shanghai'
```

```python
print(zhou.name,zhou.number,zhou.age)
```

```python
print(zhou.hometown)
```

```python
print(zhou.lovezz())
```

```python
#等价于
class Student_2():   
    name=None
    number=None
    age=None
    hometown=None
    def __init__(self,a,b,c):
        self.name=a
        self.number=b
        self.age=c
    def lovezz(self):
        print("jerrylovezhouz")

zhou=Student_2("zhou",999,18)
zhou.hometown='Shanghai'

print(zhou.name,zhou.number,zhou.age,zhou.hometown)
zhou.lovezz()
```

```python
"""
定义一个时间类 MyDate，具有年、月、日三个属性，以及一个求第二天日期的
函数和输出日期的函数。
"""

#背景知识1：

import datetime  #引入datetime库，调用一些相关函数
t=datetime.datetime.now()  #获取当前时间
print(t)
print(t.year)


```

```python
#背景知识2

"""
闰年是每隔4年出现一次，闰年的那一年是有366天

非整百年：能被4整除的为闰年。
整百年：能被400整除的是闰年。

平年365天，二月为28天。
闰年为366天，二月为29天。
"""
```

```python

"""
定义一个时间类 MyDate，具有年、月、日三个属性，以及一个求第二天日期的函数和输出日期的函数。
"""
import datetime  #引入datetime库，调用一些相关函数

class MyDate:    #定义类，满足题目要求“时间类 MyDate，具有年、月、日三个属性”
    MyYear=None
    MyMonth=None
    MyDay=None
    
    def __init__(self):    #构造方法，每次创建类的对象都会自动执行
        t=datetime.datetime.now()  #获取当前时间
        self.MyYear=t.year  #给类的栏目赋值
        self.MyMonth=t.month 
        self.MyDay=t.day
   
    def NextDate(self):    #创建类的方法，满足题目要求“求第二天日期的函数”
        a=[[31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31], [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]] 
        #创建潜逃的列表a，a[0]是平年各月份的天数，a[1]是闰年各月份的天数
        
        self.MyDay = self.MyDay + 1  #今天+1天即第二天
        
        flag = ((self.MyYear % 400==0) or (self.MyYear%4==0 and self.MyYear%100!=0)) 
        #满足被400整除或被4整除且不被100整除时，flag为1,此时a[flag]就是闰年各月份的天数
        
        if self.MyDay>a[flag][self.MyMonth - 1]:   #[self.MyMonth - 1]即闰年某月的天数(列表序号从0开始)
            
            self.MyDay = 1 #如果第二天超过了今天所在月份的最大天数，日期就变为下个月的第一天
            self.MyMonth = self.MyMonth + 1 
            
            if self.MyMonth>12: #如果下个月的月份大于12了，月份变为明年的1月
                self.MyMonth=1
                self.MyYear=self.MyYear + 1
     
    def Disp(self):       #创建类的方法，满足题目要求“输出日期的函数”
        print(self.MyYear)
        print(self.MyMonth)
        print(self.MyDay)

date=MyDate()  #创建对象

date.NextDate()
date.Disp()


    
```
